```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
tips.info()
```

```python
# Scatter plot

sns.scatterplot(data = tips, x = 'total_bill', y =  "tip", hue = "sex", style = "time", size = "size")



# Hue ->>>>>>> parameter tell you what is gender of costumer
# Style ->>>>>>>  style will tell time of visit.   Like costumer visit at lunch or dinner time
```

```python
# Replot  -> figure level  -> Square Shape

sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "sex", style = "time", size = "size")

# Hue ->>>>>>> parameter tell you what is gender of costumer
# Style ->>>>>>>  style will tell time of visit.   Like costumer visit at lunch or dinner time
```

```python
# KDE Plots
#  Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernal
# producing a continous density emtimate

tips.head()
sns.kdeplot(data=tips, x="total_bill")
```

```python
sns.displot(data=tips, x="total_bill", kind="kde")
```

```python
# hue ->  fill

sns.kdeplot(data=tips, x="total_bill", hue="sex", fill = True)
```

```python
# Rugplot

# plot marginal distribution by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individuals obsevations in an unobtrusive way.
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x ="total_bill")
```

# Matrix Plots
    *Heatmap
    *Clustermap
    

```python
# Heatmap
# Plot ractangular data as a color-encoded matrix

new_gap = gap.pivot(index="country", columns='year', values="lifeExp")
new_gap.sample(5)

# axes levelfunction
plt.figure(figsize=(15,7))
sns.heatmap(new_gap)
```

```python
new_data = gap[gap['continent'] == "Europe"].pivot(index="country", columns="year", values="lifeExp")


plt.figure(figsize = (15,7))
sns.heatmap(new_data, annot = True, linewidth = 0.5, cmap="viridis")


# annot will return the value
# LineWidth -->> This will add same gape between graph that will increase the readablity of graph

```

```python
# Clustermap

# This is very similar to heatmap

# plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to available


iris = px.data.iris()
iris.sample(5)


sns.clustermap(iris.iloc[:,[ 0,1,2,3]])
```
