```python
df = batters.head(100).sample(25, random_state = 5)
df
```

```python
plt.figure(figsize = (18, 7))
plt.title("Runs vs Strike_rate")
plt.xlabel("Runs")
plt.ylabel("Strike_rate")

plt.scatter(df["avg"], df["strike_rate"], color = "red" , s = df["runs"])

for i in range(df.shape[0]):
    plt.text(df["avg"].values[i], df["strike_rate"].values[i], df["batter"].values[i])
plt.show()
```

# 3D Surface plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize = (10,5))

ax = plt.subplot(projection = "3d")
p = ax.plot_surface(xx, yy, z, cmap = "viridis")
fig.colorbar(p)
fig.show()
```

```python
# Scatter + Surface together
```

# Contour plots

```python
z = xx**2 + yy**2
z.shape

ax1 = plt.subplot()
p = ax1.contour(xx, yy, z, cmap = "viridis")
fig.colorbar(p)

fig.show()
```

```python
z = xx**2 + yy**2
z.shape

ax1 = plt.subplot()
p = ax1.contourf(xx, yy, z, cmap = "viridis")
fig.colorbar(p)

fig.show()
```

```python
z = np.sin(xx) + np.cos(yy)
z.shape

ax1 = plt.subplot()
p = ax1.contourf(xx, yy, z, cmap = "viridis")
fig.colorbar(p)

fig.show()
```
