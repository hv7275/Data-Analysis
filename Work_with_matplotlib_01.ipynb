# Bar Chart
barchart_ver_1.jpg

    * Bivariate analysis
    * Numerical vs Categorical
    * use-case: Aggregate analysis of groups


```python
childrens = [8, 7, 5, 9,6 ]
colors = ["red", "Blue", 'Green', "Yellow", "Pink"]

plt.bar(colors, childrens, color = 'red')
plt.title("Color vs Childrens Compersion")
plt.xlabel("Colors")
plt.ylabel("Childrens")
plt.grid()
```

```python
# If you want horizonatl bar chart Then you can call "barh"
plt.barh(colors, childrens, color = "green")
plt.title("Color vs Childrens Compersion")
plt.ylabel("Colors")
plt.xlabel("Childrens")
plt.grid()
```

```python
df = pd.read_csv("/content/batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"], df["2015"], color = "yellow")
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df["2015"], width = 0.2)
plt.bar(np.arange(df.shape[0]), df["2016"],  width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df["2017"],  width = 0.2)
plt.xticks(np.arange(df.shape[0]), df["batsman"])


plt.show()
```

```python
# a problem
childrens = [8, 7, 5, 9,6 ]
colors = ["red red red red red", "Blue blue blue blue", 'Green green green', "Yellow yellow yellow", "Pink pink pink"]

plt.bar(colors, childrens, color = 'red')
plt.title("Color vs Childrens Compersion")
plt.xlabel("Colors")
plt.ylabel("Childrens")
plt.grid()


# To solve this type of problem
plt.xticks(rotation = 45)
plt.show()
```

```python
#Stacked Bar chart

plt.bar(df["batsman"], df["2015"], label = "2017")
plt.bar(df["batsman"], df["2016"], bottom= df["2017"], label = "2016")
plt.bar(df["batsman"], df["2015"], bottom = (df["2016"] + df["2017"]), label = "2017")

plt.legend()
plt.show()
```

# Pie Chart

image.png


    *Univariate/Bivariate Analysis
    *Categorical vs Numerical
    *Use-case: To find out contribution on a standard scale
    

```python
#   Simple data

data = [random.randint(1,100) for i in range(5)]
marks = ["Maths", "Python", "Stats", "Eng", "DSA"]
plt.pie(data, labels = marks)
plt.show()
```

```python
# DataSet

df = pd.read_csv("/content/gayle-175.csv")
plt.pie(df["batsman_runs"], labels= df["batsman"])
plt.show()

# But we can not see percentage


```

```python
plt.pie(df["batsman_runs"], labels= df["batsman"], autopct = "%0.1f%%")
plt.show()
```

```python
# shadow

plt.pie(df["batsman_runs"], labels= df["batsman"], autopct = "%0.1f%%", explode= [0.1, 0.1,0.1,0,0,0.1], shadow= True)
plt.show()
```
